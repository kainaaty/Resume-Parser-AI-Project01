```python
import pandas as pd

df = pd.read_csv("Resume.csv")

df.head()
```

```python
import re
import nltk

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))

def clean_resume(text):

    text = re.sub(r'[^a-zA-Z ]', '', text)

    text = text.lower()

    words = word_tokenize(text)

    filtered_words = [word for word in words if word not in stop_words]

    return " ".join(filtered_words)
```

```python
df['cleaned_resume'] = df['Resume'].apply(clean_resume)

df[['Resume', 'cleaned_resume']].head()
```

```python
print(df.columns)
```

```python
df['cleaned_resume'] = df['Resume_str'].apply(clean_resume)

df[['Resume_str', 'cleaned_resume']].head()
```

```python
skills = [
    'python',
    'java',
    'sql',
    'machine learning',
    'data analysis',
    'html',
    'css'
]
```

```python
def detect_skills(resume):

    found_skills = []

    for skill in skills:

        if skill in resume:
            found_skills.append(skill)

    return found_skills
```

```python
df['skills_found'] = df['cleaned_resume'].apply(detect_skills)

df[['skills_found']].head()
```

```python
skills = [
    'python',
    'java',
    'sql',
    'machine learning',
    'data analysis',
    'deep learning',
    'data science',
    'excel',
    'power bi',
    'communication',
    'html',
    'css',
    'javascript',
    'react',
    'c++',
    'mongodb',
    'nodejs'
]
```

```python
df['skills_found'] = df['cleaned_resume'].apply(detect_skills)

df[['skills_found']].head()
```

```python
def calculate_score(skills):

    score = len(skills)

    if score > 10:
        score = 10

    return score
```

```python
df['resume_score'] = df['skills_found'].apply(calculate_score)

df[['skills_found', 'resume_score']].head()
```

```python
top_candidates = df.sort_values(by='resume_score', ascending=False)

top_candidates[['Category', 'skills_found', 'resume_score']].head(10)
```
